```python

from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
import matplotlib.pyplot as plt
import os

# Load BLIP model and processor
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

# Function to generate captions
def generate_caption(image_path, prompt="A funny meme about "):
    image = Image.open(image_path).convert("RGB")
    inputs = processor(images=image, text=prompt, return_tensors="pt")

    out = model.generate(**inputs, max_length=50)
    caption = processor.decode(out[0], skip_special_tokens=True)
    return caption

# Display meme with caption
def display_meme(image_path, caption):
    image = Image.open(image_path)
    plt.imshow(image)
    plt.axis('off')
    plt.title(caption, fontsize=16, color='red', weight='bold')
    plt.show()

# Directory containing meme images
meme_folder = "/bin/meme"



# Loop through images and generate memes
for image_file in os.listdir(meme_folder):
    if image_file.endswith((".jpg", ".png", ".jpeg")):
        image_path = os.path.join(meme_folder, image_file)
        caption = generate_caption(image_path, prompt="A funny meme about ")
        print(f"Image: {image_file}\nCaption: {caption}\n")
        display_meme(image_path, caption)
```
